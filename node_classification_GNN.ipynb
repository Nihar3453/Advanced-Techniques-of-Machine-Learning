```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.datasets import Planetoid
from torch_geometric.transforms import NormalizeFeatures
from torch_geometric.utils import to_networkx
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report


def load_and_analyze_cora():
    dataset = Planetoid(root='data/Cora', name='Cora', transform=NormalizeFeatures())
    data = dataset[0]

    G = to_networkx(data, node_attrs=['x', 'y'], to_undirected=True)

    print("\n=== NetworkX Graph Analysis ===")
    print(f"Number of nodes: {G.number_of_nodes()}")
    print(f"Number of edges: {G.number_of_edges()}")
    print(f"Average degree: {sum(dict(G.degree()).values()) / G.number_of_nodes():.2f}")
    print(f"Is connected: {nx.is_connected(G)}")

    plt.figure(figsize=(12, 8))
    subgraph = G.subgraph(list(G.nodes())[:200])
    pos = nx.spring_layout(subgraph, seed=42)
    nx.draw(subgraph, pos, node_size=50,
            node_color=[subgraph.nodes[n]['y'] for n in subgraph.nodes()],
            cmap='Set2', with_labels=False)
    plt.title("Cora Dataset Subgraph Visualization")
    plt.show()

    return data, dataset.num_classes

class GCN(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(input_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, output_dim)
        self.dropout = torch.nn.Dropout(0.5)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.dropout(x)
        x = self.conv2(x, edge_index)
        return F.log_softmax(x, dim=1)


def train(model, data, optimizer):
    model.train()
    optimizer.zero_grad()
    out = model(data.x, data.edge_index)
    loss = F.nll_loss(out[data.train_mask], data.y[data.train_mask])
    loss.backward()
    optimizer.step()
    return loss.item()



def test(model, data, mask):
    model.eval()
    with torch.no_grad():
        out = model(data.x, data.edge_index)
        pred = out.argmax(dim=1)
        correct = pred[mask] == data.y[mask]
        acc = int(correct.sum()) / int(mask.sum())
        return acc, out



def visualize_with_networkx(model, data, epoch):
    model.eval()
    with torch.no_grad():
        out = model(data.x, data.edge_index)
        embeddings = out.cpu().numpy()

        G = to_networkx(data, to_undirected=True)
        for i, node in enumerate(G.nodes()):
            G.nodes[node]['embedding'] = embeddings[i]

        plt.figure(figsize=(14, 10))
        pos = nx.spring_layout(G, dim=2, weight=None, seed=42)
        nx.draw_networkx_nodes(G, pos, node_size=30,
                               node_color=data.y.cpu().numpy(), cmap='Set2')
        nx.draw_networkx_edges(G, pos, alpha=0.1)
        plt.title(f"Node Embeddings Visualization (Epoch {epoch})")
        plt.show()



data, num_classes = load_and_analyze_cora()

model = GCN(input_dim=data.num_features, hidden_dim=16, output_dim=num_classes)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)

best_val_acc = 0
best_test_acc = 0
train_losses, val_accs, test_accs = [], [], []

for epoch in range(1, 201):
    loss = train(model, data, optimizer)
    train_losses.append(loss)

    val_acc, _ = test(model, data, data.val_mask)
    val_accs.append(val_acc)

    test_acc, _ = test(model, data, data.test_mask)
    test_accs.append(test_acc)

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        best_test_acc = test_acc

    if epoch % 10 == 0:
        print(f'Epoch: {epoch:03d}, Loss: {loss:.4f}, Val Acc: {val_acc:.4f}, Test Acc: {test_acc:.4f}')

    if epoch in [10, 50, 100, 200]:
        visualize_with_networkx(model, data, epoch)


print(f'\n=== Final Results ===')
print(f'Best Validation Accuracy: {best_val_acc:.4f}')
print(f'Corresponding Test Accuracy: {best_test_acc:.4f}')

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')

plt.subplot(1, 2, 2)
plt.plot(val_accs, label='Validation Accuracy')
plt.plot(test_accs, label='Test Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Validation and Test Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
